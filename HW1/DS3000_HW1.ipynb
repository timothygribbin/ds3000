```python
import numpy as np
from numpy import array
from numpy import linalg

v = array([[4], [3]])
```

```python
l1_norm = linalg.norm(v, ord = 1)
print(l1_norm)
```

```python
l2_norm = linalg.norm(v, ord = 2)
print(l2_norm)
```

```python
linf_norm = linalg.norm(v, ord = np.inf)
print(linf_norm)
```

```python
l0_norm = np.count_nonzero(v)
print(l0_norm)
```
