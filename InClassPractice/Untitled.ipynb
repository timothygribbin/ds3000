```python
import numpy as np

A = np.array([[1, 3, 0], [0, 2, 3], [0, 0, 1]])

B = np.array([[1, 0, 0], [1, 4, 0], [3, 0, 2]])

M = 2 * (A @ B)

det = np.linalg.det(M)

print(det)
```

```python
A = np.array([[2, 3, 4], [0, -1, 5], [0,0,3]])

print(np.linalg.det(A))
```

```python
B = np.array([[-1, 0, 0], [2, 4, 0], [1, -3, 2]])

print(np.linalg.det(B))
```

```python
C = np.array([[-2, 0, 0], [2, 4, 0], [1, -3, 2]])

print(np.linalg.det(C))
```

```python
a = np.array([2, 2, 4])
D = np.array([[3, -1, 1], [1, 2, 0], [0, 1, 3]])
INV_D = np.linalg.inv(D)
result = a @ INV_D

print(result)
```
